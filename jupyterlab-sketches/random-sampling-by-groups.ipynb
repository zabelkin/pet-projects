Random sampling dataframe into groups
v1.0 2020-04-25

```python
import pandas as pd

# creates whatever dataframe
test_df = pd.util.testing.makeDataFrame()
```

```python
# function for sampling
# adds 'group_field' columns
# arguments: 1) _df = link to dataframe 2) _groups = list of group names
# number of items in _groups = number of samples
def random_samples_by_groups(_df, _groups = ['group0', 'group1']):
    idx = _df.index
    for sample_rows, sample_group in enumerate(_groups):
        sample_idx = _df.loc[idx,:].sample(frac=1/(len(_groups)-sample_rows)).index
        _df.loc[sample_idx, 'group_field'] =  sample_group
        idx = idx.drop(sample_idx)
```

```python
random_samples_by_groups(test_df, ['gr1','gr2','gr3','gr4', 'gr5'])
test_df.groupby('group_field').count()
```
